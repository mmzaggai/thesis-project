```python
# filter_outlier_topic.py

import pandas as pd
from pathlib import Path

# ---------------------------
# 1. Load the input file
# ---------------------------

INPUT_PATH = Path("topics_per_doc_mts30_fast_semicolon.csv")

df = pd.read_csv(INPUT_PATH, sep=";")
print("Original shape:", df.shape)

# If the column is named something else, update here:
TOPIC_COL = "topic"

# Filter
df_clean = df[df[TOPIC_COL] != -1].copy()
print("After removing topic -1:", df_clean.shape)

df_clean.head()


OUTPUT_PATH = Path("topics_per_doc_filtered.csv")
df_clean.to_csv(OUTPUT_PATH, sep=";", index=False)

print(f"Saved cleaned file → {OUTPUT_PATH.resolve()}")

```
